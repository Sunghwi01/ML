```python
titanic = pd.read_csv("../csv/train.csv")
```

```python
titanic = titanic.drop(["PassengerId", "Name", "SibSp", "Parch", "Ticket","Fare","Cabin"], axis=1)

titanic.isna().sum()
titanic['Age'].fillna(titanic["Age"].mean(), inplace=True)
titanic['Embarked'].fillna('N', inplace=True)
titanic.isna().sum()
titanic
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

encoder.fit_transform(titanic["Sex"])
label = encoder.transform(titanic["Sex"])

encoder.fit_transform(titanic["Embarked"])
labels = encoder.transform(titanic["Embarked"])

titanic["Sex"]=label
titanic["Embarked"]=labels
titanic
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
titanic_df = pd.read_csv("../csv/train.csv")

y_titanic_df = titanic_df['Survived']
X_titanic_df = titanic.drop(['Survived'],axis=1)

X_titanic_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df,test_size=0.2, random_state=110)
```

```python
dt_clf = DecisionTreeClassifier(random_state=16)

dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)

print('정확도: {0:4f}'.format(accuracy_score(y_test,dt_pred)))
```

```python
from sklearn.model_selection import KFold
import numpy as np
```

```python
## 성능 지표는 정확도(accuracy), 교차 검증 세트는 3개
scores = cross_val_score(dt_clf, X_titanic_df, y_titanic_df, scoring='accuracy', cv=12)
print('교차 검증별 정확도 : ', np.round(scores,4))
print('평균 검증 정확도:', np.round(np.mean(scores),4))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
parameters = {'max_depth' : [10], 'min_samples_split':[2,3]}
```

```python
## param_grid의 하이퍼 파라미터들을 3개의 train, test set fold로 나누어서 테스트 수행 설정
## refit = True 가 기본값임 True 이면 가장 좋은 파라미터 설정으로 재학습 시킴
grid_dt_clf = GridSearchCV(dt_clf, param_grid=parameters, cv=3, refit=True)

## 붓꽃 train 데이터로 param_grid의 하이퍼 파리미터들을 순차적으로 학습/평가
grid_dt_clf.fit(X_train, y_train)

## GridSearchCV 결과 추출하여 DataFrame으로 변환
scores_df = pd.DataFrame(grid_dt_clf.cv_results_)
scores_df
```
